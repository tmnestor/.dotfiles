```python
# Load EDA Pkgs
import pandas as pd
import random

### TESTING ONLY ###
# synthetic_data = white_list #delete after testing
# synthetic_data = black_list #delete after testing
# Shuffle Dataset


random.shuffle(synthetic_data)
# Convert to DataFrame
df = pd.DataFrame({'text':synthetic_data})
df.head(10)
```

```python
# Define Constants
ABSTAIN = -1
NOT_HOME_OFFICE = 0
HOME_OFFICE = 1
```

```python
# Import Pkgs
import re
from snorkel.labeling import labeling_function,PandasLFApplier

# Regex Lookup
@labeling_function()
def wfh_rgx_labelfn(x):
    return HOME_OFFICE if (
         any(re.search(rgx, ' '.join(x.text.split()).strip(), flags=re.I) for rgx in wfh_rgx)
         and (not any(re.search(rgx, ' '.join(x.text.split()).strip(), flags=re.I) for rgx in not_wfh_rgx))
        ) else ABSTAIN
```

```python
## Apply on Pandas
lfs = [wfh_rgx_labelfn]
applier = PandasLFApplier(lfs=lfs)
snorkel_labels = applier.apply(df=df)
```

```python
df_pred = pd.DataFrame({'prediction': snorkel_labels.flatten()})
dat1 = pd.concat([df, df_pred], axis=1)
dat1.head(10)
```
